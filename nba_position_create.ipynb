```python
import pandas as pd
import numpy as np
import time
import math

from nba_api.stats.endpoints import GameRotation
from nba_api.stats.endpoints import PlayByPlayV2
from nba_api.stats.endpoints import teamgamelogs
from nba_api.stats.endpoints import leaguedashplayerstats
from nba_api.stats.endpoints import leaguedashplayerbiostats

```

```python
#Function to create intial positions. Anybody with C is considered a big. G, G-F, F-G are considered guards. 
#Else Forwards

def consolidate_position(position):
    if 'C' in position:
        return 'big'
    elif 'G' in position:
        return 'guard'
    else:
        return 'forward'
```

```python

#Set Season
season  = '2023-24'


#Set blank data frame
master_player_df = pd.DataFrame()
```

```python

print(season)



#Iterate through NBA data using "position" as a filter, don't have a way to do this at a macro level
for position in ('G', 'F', 'C'):
    print(position)
    player_df =  leaguedashplayerstats.LeagueDashPlayerStats(season=season, \
                player_position_abbreviation_nullable = position).get_data_frames()[0]\
                [['TEAM_ABBREVIATION', 'PLAYER_ID', 'PLAYER_NAME']]


    #Bring in anthro data, height is a good checker for "wrong" positions
    antrho_df = leaguedashplayerbiostats.LeagueDashPlayerBioStats(season=season).get_data_frames()[0]\
                        [['PLAYER_ID', 'PLAYER_HEIGHT_INCHES', 'PLAYER_WEIGHT']]



    player_df = pd.merge(player_df, antrho_df, how='inner', on='PLAYER_ID')

    player_df['position'] = position
    player_df['season'] = season

    master_player_df = master_player_df.append(player_df)


    time.sleep(5)
```

```python
#Quick Check by position at a team level

pd.set_option('display.max_colwidth', None)

player_df[player_df.position == 'C']\
    .groupby(['TEAM_ABBREVIATION', 'position'])\
        .agg({ 'PLAYER_NAME':lambda x: ','.join(x)}).reset_index()
```
